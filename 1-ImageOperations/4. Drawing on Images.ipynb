## **Desenhando imagens e formas usando OpenCV**


````python
# Nossas importações e nossa função de leitura de imagem
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Define our imshow function 
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
````

Fazendo quadrados pretos

````python
#Criando uma matriz 512 x 512 com zeros, lembrando que zero no  RGB, é preto em todos os canais
image = np.zeros((512,512,3), np.uint8)

#Criando uma matriz 512 x 512 com zeros, porém com somente 1 canal, para a escala de cinza
image_gray = np.zeros((512,512), np.uint8)

# O preto deve ser o mesmo para ambas as imagens
imshow("Black Canvas - RGB Color", image)
imshow("Black Canvas - Grayscale", image_gray)
````

### **Desenhando linhas sobre o quadrado preto**

Função .line, recebe como parâmetros a imagem alvo, as coordenadas de inicio e coordenadas fim, cor  e expessura.

As coordenadas são (posição x, posição y) da matriz.

```cv2.line(image, starting cordinates, ending cordinates, color, thickness)```

````python
# Note que a figura principal é alterada com essa modificação
# Ou seja, ela não retorna uma figura nova, ela altera a mesma informada no parâmetro.
cv2.line(image, (0,0), (511,511), (255,127,0), 5)

imshow("Black Canvas With Diagonal Line", image)
````

### **Desenhando retangulos**

Função .rectangle recebe como parâmetros: a imagem alvo, vértice inicial (ponto superior esquerdo), vértice opositor (ponto inferior direito), cor, expessura

```cv2.rectangle(image, starting vertex, opposite vertex, color, thickness)```

````python
# criamos um novo quadrado preto
image = np.zeros((512,512,3), np.uint8)

# Com expessura positiva, temos uma linha de borda
cv2.rectangle(image, (100,150), (300,250), (127,50,127), 5)
imshow("Black Canvas With Pink Rectangle", image)
````

````python
# criamos um novo quadrado preto
image = np.zeros((512,512,3), np.uint8)

# Com expessura negativa, preenchemos o quadrado todo
cv2.rectangle(image, (100,150), (300,250), (127,50,127), -5)
imshow("Black Canvas With Pink Rectangle", image)
````

### **E como fazemos circulos?**

Função .circle recebe: imagem, o ponto central,  o raio do círculo, cor, preenchimento

```cv2.circle(image, center, radius, color, fill)```

````python
image = np.zeros((512,512,3), np.uint8)

cv2.circle(image, (350, 350), 100, (15,150,50), 1) 
imshow("Black Canvas With Green Circle", image)
````

### **Desenhando Polígonos**

```cv2.polylines(image, points, Closed?, color, thickness)```

image:
A imagem onde as linhas poligonais serão desenhadas.

[points]:
Uma lista de pontos que formam os vértices do polígono. Os pontos devem estar em um array NumPy, como np.array([[x1, y1], [x2, y2], ...], dtype=np.int32).
É importante colocar os pontos entre colchetes [], mesmo que seja um único conjunto de pontos, para que o OpenCV entenda como uma lista de contornos.

isClosed:
Um valor booleano (True ou False) que indica se o polígono deve ser fechado automaticamente (ou seja, se a última linha deve conectar o último ponto ao primeiro).

color:
A cor da linha, fornecida como uma tupla (B, G, R) para imagens coloridas.

thickness:
Define a espessura da linha.



````python
image = np.zeros((512,512,3), np.uint8)

# Definindo 4 pontos
pts = np.array( [[10,50], [400,50], [90,200], [50,500]], np.int32)
print(pts.shape)
pts = pts.reshape((-1,1,2))
#Necessário somente, se, voce utilizar os pontos para demais funções do opencv.
#Funciona sem.

print(pts.shape)

cv2.polylines(image, [pts], True, (0,0,255), 3)
imshow("Black Canvas with Red Polygon", image)
````

### **Adicionando escrita na imagem**

função .putText: imagem, texto a ser mostrado, ponto mais a esquerda de inicio, a font, tamanho da font,  cor, expessura

```cv2.putText(image, 'Text to Display', bottom left starting point, Font, Font Size, Color, Thickness)```

**Fontes disponiveis**
- FONT_HERSHEY_SIMPLEX, FONT_HERSHEY_PLAIN
- FONT_HERSHEY_DUPLEX,FONT_HERSHEY_COMPLEX 
- FONT_HERSHEY_TRIPLEX, FONT_HERSHEY_COMPLEX_SMALL
- FONT_HERSHEY_SCRIPT_SIMPLEX
- FONT_HERSHEY_SCRIPT_COMPLEX

````python
image = np.zeros((1000,1000,3), np.uint8)
ourString =  'AMO A MARIA EDUARDA'
cv2.putText(image, ourString, (155,290), cv2.FONT_HERSHEY_PLAIN, 4, (40,200,0), 4)

cv2.putText(image, ourString, (155,390), cv2.FONT_HERSHEY_PLAIN, 4, (40,200,0), 4)
imshow("Imagem com texto", image)
````
