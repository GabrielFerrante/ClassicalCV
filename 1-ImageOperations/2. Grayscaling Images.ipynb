# **Grayscaling Images**

Iremos aprender:
1. Converter uma imagem em escala de cinza
2. Ver as diferenças de dimensões entre imagens coloridas e de cinza



```python
import cv2
from matplotlib import pyplot as plt
```

```python
# Vamos dar um foco a mais na figura
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h #Proporção de tela
    plt.figure(figsize=(size * aspect_ratio,size)) #10 vezes a proporção, mantendo a resolução
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
# Chamando nossa função que criamos 
image = cv2.imread('PatricioGatoNenudo.jpg')

imshow("PATRICIO", image)
```

```python
#Mesma função, só que diferente.

image.shape[:2] # (1200, 1920) (height, width)
def imshow(title = "", image = None, size = 10):

    h, w = image.shape[0], image.shape[1]
    aspect_ratio = w/h
 
    plt.figure(figsize=(size * aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()

imshow("Castara, Tobago", image)
```

```python
# Função cvtColor, passando a imagem e qual espaço de cor você converter
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

imshow("Imagem em Cinza, convertida", gray_image)
```
