### **Carregando uma imagem em uma variavel**

```python
#função imread 
image = cv2.imread('./images/castara.jpeg')
```

### **Apresentando a imagem com a biblioteca Matplotlib**

```python
from matplotlib import pyplot as plt

#Função imshow apresenta a imagem
#Função cvtColor converte a imagem para um espaço de cor para outro
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

Vamos criar uma função que apresente uma imagem com um título

```python
def imshow(title = "", image = None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title) #função title do matplotlib
    plt.show()
```

```python

imshow("Imagem da praia", image)
```

### **Salvando Imagens**

```python
# Função imwrite para salvar a imagem no diretório atual deste arquivo notebook
cv2.imwrite('output.jpg', image)
```

```python
#podemos salvar também em PNG (Portable Network Graphics) que é um formato de imagem bitmap sem perdas
cv2.imwrite('output.png', image)
```

### **Descobrindo as dimensões de uma imagem**

Não se esqueça, imagens são Arrays multidimensionais:

uma imagem essencialmente é 2D, que diz respeito a quantidade de pixels no eixo X e Y, ou seja, sua resolução.

O canal de cores é a terceira dimensão de uma imagem digital, que vai contar com um valor em cada uma das 3 cores do RGB, por exemplo.
A cor do pixel é a mistura do brilho de vermelho, verde e azul (cores primárias) naquela posição do grid da imagem.

![](https://raw.githubusercontent.com/rajeevratan84/ModernComputerVision/main/array.png?token=ADLZD2HNEL33JAKTYRM3B5C7WMIV4)



Para isso, precisamos de um mecanismo para ler arrays multidimensionais de maneira fácil. 
Numpy is POWER.

```python
import numpy as np

print(image.shape)
```

```python
# To access a dimension, simply index it by using 0, 1 or 2.
image.shape[0]
```

O primeiro valor da lista .shape, é a altura, no índice 0

O segundo valor, é a largura da imagem, no índice 1

e ao final, é o número de cores do canal (RGB)


```python
print('Altura da Imagem: {} pixels'.format(int(image.shape[0])))
print('Largura da Imagem: {} pixels'.format(int(image.shape[1])))
print('Profundidade da Imagem: {} Componentes de cores'.format(int(image.shape[2])))
```
